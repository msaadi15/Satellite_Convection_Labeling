```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# --- config: exact suffixes in your DataFrame ---
SUFFIX_A = "_OpenCV_vs_Expert"
SUFFIX_B = "_UNet_vs_Expert"

# --- infer metric base names robustly from columns ---
def infer_metrics_from_columns(cols, suf_a=SUFFIX_A, suf_b=SUFFIX_B):
    metrics = set()
    for c in cols:
        if c.endswith(suf_a):
            metrics.add(c[:-len(suf_a)])
        elif c.endswith(suf_b):
            metrics.add(c[:-len(suf_b)])
    return sorted(metrics)

# --- paired comparison and statistics for a single metric ---
def compare_metric(df, metric, suf_a=SUFFIX_A, suf_b=SUFFIX_B, bootstrap_iters=2000):
    col_a = f"{metric}{suf_a}"
    col_b = f"{metric}{suf_b}"
    # drop rows with NaN in either column for paired comparison
    pair = df[[col_a, col_b]].dropna()
    a = pair[col_a].values
    b = pair[col_b].values
    diff = a - b
    n = len(diff)
    if n == 0:
        return None  # no data
    
    # counts
    a_better = int(np.sum(a > b))
    b_better = int(np.sum(b > a))
    equal = int(np.sum(a == b))
    
    # descriptive
    desc = {
        f"{metric}{suf_a}_mean": float(np.mean(a)),
        f"{metric}{suf_b}_mean": float(np.mean(b)),
        "mean_diff": float(np.mean(diff)),
        "median_diff": float(np.median(diff)),
        "std_diff": float(np.std(diff, ddof=1)),
        "n": n
    }
    
    # paired t-test
    ttest = stats.ttest_rel(a, b, nan_policy='omit')
    
    # Wilcoxon signed-rank (catch exceptions)
    try:
        wilcoxon = stats.wilcoxon(a, b, alternative='two-sided')
    except Exception:
        wilcoxon = None
    
    # Cohen's d (paired)
    cohen_d = float(np.mean(diff) / (np.std(diff, ddof=1) + 1e-12))
    
    # bootstrap CI for mean diff
    rng = np.random.default_rng()
    boot_means = []
    for _ in range(bootstrap_iters):
        samp = rng.choice(diff, size=n, replace=True)
        boot_means.append(np.mean(samp))
    lower = float(np.percentile(boot_means, 2.5))
    upper = float(np.percentile(boot_means, 97.5))
    
    return {
        "metric": metric,
        "col_a": col_a, "col_b": col_b,
        "OpenCV_better_count": a_better,
        "UNet_better_count": b_better,
        "Equal_count": equal,
        "desc": desc,
        "t_test_stat": float(ttest.statistic), "t_test_p": float(ttest.pvalue),
        "wilcoxon": wilcoxon,
        "cohen_d": cohen_d,
        "mean_diff_CI": (lower, upper)
    }

# --- Bland-Altman plotting helper ---
def bland_altman_plot(a, b, ax=None, title=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(6,4))
    mean = (a + b) / 2.0
    diff = a - b
    md = np.mean(diff)
    sd = np.std(diff, ddof=1)
    loa_upper = md + 1.96 * sd
    loa_lower = md - 1.96 * sd
    
    ax.scatter(mean, diff, alpha=0.6)
    ax.axhline(md, linestyle='--', label=f"mean diff={md:.3f}")
    ax.axhline(loa_upper, linestyle='--', label=f"+1.96sd={loa_upper:.3f}")
    ax.axhline(loa_lower, linestyle='--', label=f"-1.96sd={loa_lower:.3f}")
    ax.set_xlabel('Mean of OpenCV & UNet')
    ax.set_ylabel('Difference (OpenCV - UNet)')
    if title:
        ax.set_title(title)
    ax.legend()
    return ax

# --- Main routine: compute summary for all detected metrics ---
def run_all_comparisons(df, suf_a=SUFFIX_A, suf_b=SUFFIX_B, bootstrap_iters=2000, plot=True):
    metrics = infer_metrics_from_columns(df.columns, suf_a, suf_b)
    results = []
    for m in metrics:
        col_a = f"{m}{suf_a}"
        col_b = f"{m}{suf_b}"
        if col_a not in df.columns or col_b not in df.columns:
            continue
        res = compare_metric(df, m, suf_a, suf_b, bootstrap_iters=bootstrap_iters)
        if res:
            results.append(res)
    # build result DataFrame
    rows = []
    for r in results:
        d = r["desc"]
        rows.append({
            "Metric": r["metric"],
            f"OpenCV_mean": d[f"{r['metric']}{suf_a}_mean"],
            f"UNet_mean": d[f"{r['metric']}{suf_b}_mean"],
            "OpenCV_better_count": r["OpenCV_better_count"],
            "UNet_better_count": r["UNet_better_count"],
            "Equal_count": r["Equal_count"],
            "Mean_diff": d["mean_diff"],
            "Mean_diff_CI_low": r["mean_diff_CI"][0],
            "Mean_diff_CI_high": r["mean_diff_CI"][1],
            "Cohen_d": r["cohen_d"],
            "t_stat": r["t_test_stat"],
            "t_pvalue": r["t_test_p"]
        })
    df_summary = pd.DataFrame(rows)
    
    if plot and not df_summary.empty:
        for m in df_summary["Metric"].tolist():
            col_a = f"{m}{suf_a}"
            col_b = f"{m}{suf_b}"
            pair = df[[col_a, col_b]].dropna()
            a = pair[col_a].values
            b = pair[col_b].values
            
            # boxplot side-by-side using seaborn
            melt = pair.rename(columns={col_a: "OpenCV", col_b: "UNet"}).melt(var_name="Method", value_name="Value")
            plt.figure(figsize=(6,4))
            sns.boxplot(data=melt, x="Method", y="Value", palette="Set2")
            plt.title(f"Boîtes à moustaches du score {m} sur les images de test")  # French title
            plt.tight_layout()
            plt.show()
            
            # Bland-Altman
            fig, ax = plt.subplots(figsize=(6,4))
            bland_altman_plot(a, b, ax=ax, title=f"{m} Bland-Altman (OpenCV vs UNet)")
            plt.tight_layout()
            plt.show()
            
    return df_summary

# --- Example usage ---
# df_metrics should already exist in your environment
df_summary = run_all_comparisons(df_metrics, bootstrap_iters=2000, plot=True)
```
